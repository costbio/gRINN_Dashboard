```python
from dash import Dash, dcc, html
import py3Dmol 
```

```python
app = Dash()
```

```python
pdb_file = r"/home/tugba/grinn_workflow_results_new_0/system_dry.pdb"
```

```python
with open (pdb_file) as file:
    pdb_data = file.read()
```

```python
def create_3d_view(pdb_data):
    view = py3Dmol.view(width=800, height=400)
    view.addModel(pdb_data, "pdb")
    view.setStyle({'stick': {}})
    view.zoomTo()
    return view
```

```python
# Dash layout'u oluşturuyoruz
app.layout = html.Div([
    html.H1("Protein 3D Visualization", style={'textAlign': 'center'}),
    
    # 3Dmol görselleştirmesi
    html.Div(children=[
        html.H3("3D Protein Structure", style={'textAlign': 'center'}),
        dcc.Graph(
            id='3d-protein-view',
            figure=create_3d_view(pdb_data).get_figure()
        )
    ])
])
```

```python
if __name__ == "__main__":

    app.run(debug = True)
```
